```python
# Load libraries

import nibabel as nib
from nilearn import plotting, image
import numpy as np
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore", category=FutureWarning)
```

```python
# Load data

fmri_path = 'data/sub-01_ses-1_task-rest_bold_space-MNI152NLin2009cAsym_preproc.nii.gz'
fmri = image.load_img(fmri_path)
```

```python
# Print dimensions and type of data
print(type(fmri))
print(fmri.shape)

fmri_arr = fmri.get_data()

print(type(fmri_arr))
print(fmri_arr.shape)
```

```python
# Select one volume (image.index_img) 
fmri_volume = image.index_img(fmri, 20)
print(type(fmri_volume), fmri_volume.shape)
print(type(fmri), fmri.shape)
print(fmri==fmri_volume)
```

```python
# Make a plot (plotting.plot_epi)

plotting.plot_epi(fmri_volume, cmap='gray') # echo planner imaging

```

```python
# Convert nibabel file to numpy array (get_data)

fmri_arr = fmri.get_data()


# Check type and dimensions

print(type(fmri_arr), fmri_arr.shape)
```

```python
# Plot timeseries of selected voxels

voxels = fmri_arr[25,33,32, :], fmri_arr[13,14,29, :]
#print(voxels_coors)s

for i, voxel in enumerate(voxels):
    plt.plot(voxel, label = 'Voxel %i'%i)
plt.legend()
plt.show()
```

```python
# Generate random timeseries with the length of fMRI data (np.random.uniform (low, high, size=(,))

random_voxel = np.random.uniform(min(voxels[0]), max(voxels[1]), len(voxels[0]))

plt.plot(random_voxel)
plt.show()
```
